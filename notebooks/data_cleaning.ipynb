```python
def info(x):
    n_missing = x.isnull().sum().sort_values(ascending=False)
    p_missing = (x.isnull().sum()/x.isnull().count()).sort_values(ascending=False)
    dtype = x.dtypes
    count = x.count()
    missing_ = pd.concat([n_missing, dtype, count],axis=1, keys = [
        'number_missing',
        'type',
        'count'
    ])
    return missing_
```

```python
info(df)
```
